##**Average Treatment Effect**

```python
df_experiment['total_score'] = df_experiment['Ad_Engagement'] + df_experiment['Purchase_Likelihood'] + df_experiment['Visual_Appeal'] + df_experiment['Message_Clarity'] + df_experiment['Overall_Impression'] + df_experiment['Effort_Level'] + df_experiment['Style_Enjoyment']
```

```python
#Calculating the Average Treatment Effect (ATE) for the 'total_score' variable based treatment group ('Treatment_Group')
mean_total_score_control = df_experiment[df_experiment['Treatment_Group'] == 0]['total_score'].mean()
mean_total_score_treatment = df_experiment[df_experiment['Treatment_Group'] == 1]['total_score'].mean()
ATE_total_score = (mean_total_score_control - mean_total_score_treatment)
print(ATE_total_score)
```

Positive value of ATE indicates an increase in the outcome variable due to the treatment.

```python
import statsmodels.api as sm
from stargazer.stargazer import Stargazer

regression_model = sm.OLS.from_formula('total_score ~ Treatment_Group', data=df_experiment).fit(cov_type='HC3')
Stargazer([regression_model])
```

Our analysis reveals that participants exposed to black and white ads (Treatment Group) scored 1.669 points lower on the total score compared to those who saw color ads (Control Group). This effect is not statistically significant (p-value < 0.1).

However, this result isn't yet conclusive (p-value < 0.1). A larger sample size might be needed to solidify the finding as statistically significant (typically requiring a p-value below 0.05).
In simpler terms, there's a hint that black and white ads might be linked to lower scores, but more data is needed to be certain.

The R-squared value is 0.032, indicating that the independent variable 'Treatment_Group' explains approximately 3.2% of the variance in 'total_score'.

The residual error is 4.650, indicating the typical amount of error or variability in the model's predictions.


```python
gender_model = sm.OLS.from_formula('total_score ~ Treatment_Group + Gender', data=df_experiment).fit(cov_type='HC3')

Stargazer([regression_model, gender_model])
```

The decrease in standard error and increase in R-squared after controlling for the gender covariate indicate that the model has become more precise and explanatory. This suggests that the inclusion of the gender covariate has improved the accuracy and comprehensiveness of the regression analysis.

Additionally, the inclusion of gender as a covariate did not substantially change the treatment effect estimate, but it increased the precision of the estimate, resulting in a more precise treatment effect estimate compared to the previous regression.

```python
ad_engagement_model = sm.OLS.from_formula('total_score ~ Treatment_Group + Ad_Engagement', data=df_experiment).fit(cov_type='HC3')
purchase_likelihood_model = sm.OLS.from_formula('total_score ~ Treatment_Group + Purchase_Likelihood', data=df_experiment).fit(cov_type='HC3')
visual_appeal_model = sm.OLS.from_formula('total_score ~ Treatment_Group + Visual_Appeal', data=df_experiment).fit(cov_type='HC3')
message_clarity_model = sm.OLS.from_formula('total_score ~ Treatment_Group + Message_Clarity', data=df_experiment).fit(cov_type='HC3')
overall_impression_model = sm.OLS.from_formula('total_score ~ Treatment_Group + Overall_Impression', data=df_experiment).fit(cov_type='HC3')
effort_level_model = sm.OLS.from_formula('total_score ~ Treatment_Group + Effort_Level', data=df_experiment).fit(cov_type='HC3')
style_enjoyment_model = sm.OLS.from_formula('total_score ~ Treatment_Group + Style_Enjoyment', data=df_experiment).fit(cov_type='HC3')

Stargazer([regression_model, ad_engagement_model, visual_appeal_model, message_clarity_model, overall_impression_model, effort_level_model, style_enjoyment_model])
```

The regression analysis identified Overall Impression (-0.817), Style Enjoyment (-0.792), and Visual Appeal (0.903) as robust control variables. Their coefficients closely mirrored the treatment group coefficient (-1.669). Effective control variables minimally influence the treatment effect estimate.

This congruence suggests these variables effectively controlled for perception-related factors, exhibiting a comparable impact on the total score as the treatment (black and white ads). Conversely, variables with coefficients substantially deviating from -1.669 might not be ideal controls and could potentially confound the analysis.

In essence, the analysis accounted for how participants perceived the ad's style and overall look (independent of color). These factors significantly impacted the total score and served as effective controls becau

In simpler terms, seeing black and white ads is linked to lower scores, and this effect holds true even when we consider how people felt about the ad's style. Additionally, including these style-related factors in the analysis provides a more complete picture of what influences total score.

##**Balance Check**

```python
from scipy.stats import chi2_contingency

contingency_table_gender = pd.crosstab(df_experiment['Treatment_Group'], df_experiment['Gender'])
chi2_gender, pvalue_gender, _, _ = chi2_contingency(contingency_table_gender)
print(f"Chi-square statistic for Gender (chi2): {chi2_gender}")
print(f"P-value for Gender (p): {pvalue_gender}")

contingency_table_country = pd.crosstab(df_experiment['Treatment_Group'], df_experiment['Country'])
chi2_country, pvalue_country, _, _ = chi2_contingency(contingency_table_country)
print(f"Chi-square statistic for Country (chi2): {chi2_country}")
print(f"P-value for Country (p): {pvalue_country}")
```

**Gender:** There's likely no statistically significant association between the variable being analyzed (possibly ad preference, click-through rate, etc.) and gender.

**Country:** There might be a weak association between the analyzed variable and country. However, with a p-value slightly above 0.05, we can't definitively claim statistical significance. A larger sample size or a more stringent significance level might reveal a clearer picture.

```python
from statsmodels.stats.proportion import proportions_ztest

count = len(df_experiment[df_experiment['Treatment_Group'] == 1])
nobs = 119
value = 0.5

stat, pval = proportions_ztest(count, nobs, value)
print('{0:0.3f}'.format(pval))
```
